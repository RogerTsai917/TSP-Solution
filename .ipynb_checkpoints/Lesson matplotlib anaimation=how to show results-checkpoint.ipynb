# Step 2: Draw a gridworld

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

fig, ax =plt.subplots(figsize=(4,3))
xs = np.linspace(0,5,5)
ys = np.linspace(0,4,4)
ax = plt.gca()
# grid "shades" (boxes)
w, h = xs[1] - xs[0], ys[1] - ys[0]
for i, x in enumerate(xs[:-1]):
    for j, y in enumerate(ys[:-1]):
        if i % 2 == j % 2: # racing flag style
            ax.add_patch(Rectangle((x, y), w, h, fill=True, color='#008610', alpha=.1))
# grid lines
for x in xs:
    plt.plot([x, x], [ys[0], ys[-1]], color='black', alpha=.33, linestyle=':')
for y in ys:
    plt.plot([xs[0], xs[-1]], [y, y], color='black', alpha=.33, linestyle=':')

# grid points
    #circle=plt.Circle((1,1),0.5)
plt.show()
```

# testing

```python
import matplotlib
matplotlib.use('Qt4Agg')

import numpy as np
from matplotlib import pyplot
from matplotlib import animation
from math import sin

data = np.zeros((1000,4))

data[:,0] = [20*(1+sin(float(x)/200)) for x in range(1000)]
data[:,1] = [20*(1+sin(float(x)/100)) for x in range(1000)]
data[:,2] = [20*(1+sin(float(x)/50)) for x in range(1000)]
data[:,3] = [20*(1+sin(float(x)/25)) for x in range(1000)]

fig=pyplot.figure()
ax = pyplot.axes(xlim=(0, 40), ylim=(0, 40))

circle1=pyplot.Circle((data[0,0],1.0),0.2,fc='y')
circle2=pyplot.Circle((data[0,1],1.0),0.2,fc='g')
circle3=pyplot.Circle((data[0,2],1.0),0.2,fc='r')
circle4=pyplot.Circle((data[0,3],1.0),0.2,fc='b')

def init():
    circle1.center=(data[0,0],1)
    circle2.center=(data[0,1],1) 
    circle3.center=(data[0,2],1)
    circle4.center=(data[0,3],1)
    ax.add_patch(circle1)
    ax.add_patch(circle2)
    ax.add_patch(circle3)
    ax.add_patch(circle4)
    return circle1, circle2, circle3, circle4

def animate(i):
    # for state in data:
    circle1.center=(data[i,0],1)
    circle2.center=(data[i,1],1)
    circle3.center=(data[i,2],1)
    circle4.center=(data[i,3],1)
    return circle1, circle2, circle3, circle4


anim=animation.FuncAnimation(fig,animate,init_func=init,frames=1000,blit=True)

pyplot.show()

from IPython.display import HTML
HTML(ani.to_jshtml())
```

# Step 8 test circle animation

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
#matplotlib inline
testData = np.random.rand(100,2)-[0.5, 0.5]

fig, ax = plt.subplots()
def init():
    fig.clf()
    sctPlot, = ax.plot(testData[:,0], testData[:,1], ".")

# i is the radius
def animate(i):
    init()
    q = 20 #initial index for the starting position
    pos = testData[q,:]
    circle1=plt.Circle(pos,i,color='r',fill=False, clip_on = False)
    fig.gca().add_artist(circle1)

def init():
    sctPlot, = ax.plot(testData[:,0], testData[:,1], ".")
    return sctPlot,

ani = animation.FuncAnimation(fig, animate, np.arange(0.4, 2, 0.1), init_func=init,
    interval=25, blit=False)
plt.show()

from IPython.display import HTML
HTML(ani.to_jshtml())
```

# Step  9: Working to save to gif frames (OK)

```python

import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#install JSAanimate if not installed
!pip install JSAnimation
!pip install IPython

from JSAnimation.IPython_display import display_animation
import matplotlib.animation
from IPython.display import display


'''
# Plot the graph
def plotgraph(filename):
    plt.figure()
    predicted = model(torch.from_numpy(x_train)).detach().numpy()
    plt.plot(x_train, y_train, 'ro', label='Original data')
    plt.plot(x_train, predicted, label='Fitted line')
    plt.legend()

# save graph for replay
def savegraph(filename):
    predicted = model(torch.from_numpy(x_train)).detach().numpy()
    plt.figure()
    plt.plot(x_train, y_train, 'ro', label='Original data')
    plt.plot(x_train, predicted, label='Fitted line')
    plt.legend()
    plt.savefig(filename)
    plt.close()


frames=[]
fig_no =0
def display_frames_as_gif(frames):
    plt.figure(figsize(5,5))
    patch=plt.imshow(frames[0])
    plt.axis('off')

def animate(i):
    patch.set_data(frame[i])
'''

history=[]

# Hyper-parameters
input_size = 1
output_size = 1
num_epochs = 60
learning_rate = 0.001

# Toy dataset
x_train = np.array([[3.3], [4.4], [5.5], [6.71], [6.93], [4.168], 
                    [9.779], [6.182], [7.59], [2.167], [7.042], 
                    [10.791], [5.313], [7.997], [3.1]], dtype=np.float32)



y_train =np.array([[1.7], [2.76], [2.09], [3.19], [1.694], [1.573], 
                    [3.366], [2.596], [2.53], [1.221], [2.827], 
                    [3.465], [1.65], [2.904], [1.3]], dtype=np.float32)


#x_train = np.concatenate([x_train,x_train+2.5,x_train-2,x_train+4])
#y_train = np.concatenate([y_train,y_train+2.7,y_train-1,y_train+6])
# Linear regression model
model = nn.Linear(input_size, output_size)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)  

# Train the model
for epoch in range(num_epochs):
    # Convert numpy arrays to torch tensors
    inputs = torch.from_numpy(x_train)
    targets = torch.from_numpy(y_train)

    # Forward pass
    outputs = model(inputs)
    loss = criterion(outputs, targets)
    
    # Backward and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 3 == 0:
        print ('Epoch [{}/{}], Loss: {:.4f}'.format(epoch+1, num_epochs, loss.item()))
        predicted = model(torch.from_numpy(x_train)).detach().numpy()
        history.append(predicted)
        

# Save the model checkpoint



fig, ax = plt.subplots()
ax.plot(x_train, y_train,'ro', label='Original data')

line, =ax.plot(x_train,history[0])
#line, =ax.legend()
    
def init():
    line.set_ydata(history[0])
    return line,

def animate(i):
    line.set_ydata(history[i])
    return line,





ani = matplotlib.animation.FuncAnimation(fig, func=animate, init_func=init, interval=200, frames=len(history),blit=False)
#plt.show()

from IPython.display import HTML
HTML(ani.to_jshtml())
```
